```python
import cv2
import numpy as np
import pandas as pd
import pickle as pkl
import os
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def get_image_metadata(img_path):
    """
    Load image and return metadata, including
    - image size, as separate height and width values
    - image mode
    - image format
    """
    img = Image.open(img_path)
    return {
        'height': img.height,
        'width': img.width,
        'mode': img.mode,
        'format': img.format
    }
```

```python
from PIL import UnidentifiedImageError


TRAIN_IMG_PATH = "./pix2tex/dataset/formulae/train/"

# Get metadata for all images in the training set
train_img_metadata = []

for img_path in tqdm(os.listdir(TRAIN_IMG_PATH)):
    try:
        dic = {
            "image": img_path
        }
        dic.update(get_image_metadata(os.path.join(TRAIN_IMG_PATH, img_path)))

        train_img_metadata.append(dic)
    except UnidentifiedImageError:
        print(f"Error identifying image: {img_path}")

train_img_metadata = pd.DataFrame(train_img_metadata)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].hist(train_img_metadata['height'], bins=50, )
ax[0].set_title('Height distribution')
ax[0].set_xlabel('Height')
ax[0].set_ylabel('Frequency')

ax[1].hist(train_img_metadata['width'], bins=50)
ax[1].set_title('Width distribution')
ax[1].set_xlabel('Width')
ax[1].set_ylabel('Frequency')
```

# Compare image total sizes

```python
# create the feature "dimensions", a tuple of height and width
train_img_metadata['dimensions'] = list(zip(train_img_metadata['height'], train_img_metadata['width']))
multiline_train_img_metadata['dimensions'] = list(zip(multiline_train_img_metadata['height'], multiline_train_img_metadata['width']))
handwritten_train_img_metadata['dimensions'] = list(zip(handwritten_train_img_metadata['height'], handwritten_train_img_metadata['width']))
```

```python
# compare the value counts of each dimension in each dataset
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

train_img_metadata['dimensions'].value_counts().plot(kind='barh', ax=ax[0])
ax[0].set_title('Single-line dataset')
ax[0].set_xlabel('Dimensions')

multiline_train_img_metadata['dimensions'].value_counts().plot(kind='barh', ax=ax[1])
ax[1].set_title('Multiline dataset')
ax[1].set_xlabel('Dimensions')

handwritten_train_img_metadata['dimensions'].value_counts().plot(kind='barh', ax=ax[2])
ax[2].set_title('Handwritten dataset')
ax[2].set_xlabel('Dimensions')

```

```python
# Calculate cumulative percentage of value counts
train_cumulative_perc = train_img_metadata['dimensions'].value_counts(normalize=True).sort_index().cumsum()
multiline_train_cumulative_perc = multiline_train_img_metadata['dimensions'].value_counts(normalize=True).sort_index().cumsum()
handwritten_train_cumulative_perc = handwritten_train_img_metadata['dimensions'].value_counts(normalize=True).sort_index().cumsum()

# Filter dimensions that account for 80% of the total number of observations
train_top_dimensions = train_cumulative_perc[train_cumulative_perc <= 0.8].index
multiline_train_top_dimensions = multiline_train_cumulative_perc[multiline_train_cumulative_perc <= 0.8].index
handwritten_train_top_dimensions = handwritten_train_cumulative_perc[handwritten_train_cumulative_perc <= 0.8].index

# Plot
fig, ax = plt.subplots(3, 1, figsize=(10, 30))

train_cumulative_perc[train_cumulative_perc.index.isin(train_top_dimensions)].plot(kind='barh', ax=ax[0])
ax[0].set_title('Single-line dataset')
ax[0].set_xlabel('Cumulative Percentage')

multiline_train_cumulative_perc[multiline_train_cumulative_perc.index.isin(multiline_train_top_dimensions)].plot(kind='barh', ax=ax[1])
ax[1].set_title('Multiline dataset')
ax[1].set_xlabel('Cumulative Percentage')

handwritten_train_cumulative_perc[handwritten_train_cumulative_perc.index.isin(handwritten_train_top_dimensions)].plot(kind='barh', ax=ax[2])
ax[2].set_title('Handwritten dataset')
ax[2].set_xlabel('Cumulative Percentage')

```

# Resize multiline and handwritten images

```python
from pix2tex.cli import minmax_size
```

```python
def resize_image(img_path, target_height):
    """
    Load image from the given path and resize it while maintaining the aspect ratio.
    The height of the resized image will be equal to the target_height.
    """
    img = Image.open(img_path)
    width, height = img.size
    aspect_ratio = height / width
    target_width = int(target_height / aspect_ratio)
    resized_img = img.resize((target_width, target_height))
    return resized_img
```

```python
def resize_img_multiline(img_path):
    """
    Resize to meet multline YAML config parameters
    """
    # max_height: 192
    # max_width: 672
    # min_height: 32
    # min_width: 32
    img = Image.open(img_path)
    img = minmax_size(img, max_dimensions=(672, 192), min_dimensions=(32, 32))
    return img
```

```python
HANDWRITTEN_RESIZED_TRAIN_IMG_PATH = "./pix2tex/dataset/handwritten/train_resized/"

# Create the directory if it does not exist
if not os.path.exists(HANDWRITTEN_RESIZED_TRAIN_IMG_PATH):
    os.makedirs(HANDWRITTEN_RESIZED_TRAIN_IMG_PATH)

# Resize all images in the training set
for img_path in tqdm(os.listdir(HANDWRITTEN_TRAIN_IMG_PATH)):
    try:
        resized_img = resize_img_multiline(os.path.join(HANDWRITTEN_TRAIN_IMG_PATH, img_path))
        resized_img.save(os.path.join(HANDWRITTEN_RESIZED_TRAIN_IMG_PATH, img_path))
    except UnidentifiedImageError:
        print(f"Error resizing image: {img_path}")
```
